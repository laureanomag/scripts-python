**Dibujar figuras dentro de una imagen**

```python
# Importar librerías y verificar instalación
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print(cv2.__version__)
```

```python
# Crear una imagen de color negro
```

```python
imagen = np.zeros(shape=(500, 500, 3), dtype=np.uint8)
```

```python
print(imagen.shape)
```

```python
plt.imshow(imagen)
```

EJERCICIO 3


```python
# Dibujar un rectángulo de color verde
# Color verde en BGR: (0, 255, 0)
cv2.rectangle(imagen, pt1=(30, 290), pt2=(170, 370), color=(0, 255, 0), thickness=10)

```

```python
#Dibujar el círculo magenta (derecha)
# Color magenta en BGR: (255, 0, 255)
cv2.circle(imagen, center=(300, 150), radius=75, color=(255, 0, 255), thickness=10)
```

```python
# Dibujar la línea vertical blanca (extremo derecho)
# Color blanco en BGR: (255, 255, 255)
cv2.line(imagen, pt1=(430, 40), pt2=(430, 450), color=(255, 255, 255), thickness=10)
```

EJERCICIO 6


```python
imagen = np.zeros(shape=(500, 500, 3), dtype=np.uint8)
```

```python
# Naranja (BGR)
cv2.rectangle(imagen, pt1=(40, 160), pt2=(110, 360), color=(0, 165, 255), thickness=10)

# Convertimos la imagen para que Matplotlib la entienda bien
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
# usamos la variable convertida
plt.imshow(imagen_rgb)
plt.axis("off")
plt.show()
```

```python
#Dibujar un círculo verde
# center: coordenadas del centro (x, y)
# radius: radio del círculo
cv2.circle(imagen, center=(320, 300), radius=90, color=(0, 255, 0), thickness=10)

# Convertir y mostrar
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
plt.imshow(imagen_rgb)
plt.axis("off")
plt.show()
```

```python
# 3. Dibujar una recta rosa

cv2.line(imagen, pt1=(150, 60), pt2=(470, 60), color=(215, 100, 255), thickness=10)

# Convertir y mostrar el resultado final con todas las figuras
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB);
plt.imshow(imagen_rgb);
plt.axis("off");
plt.show()
```
